```python
import cv2
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
import requests
print('Start Downloading')
url = 'http://1.bp.blogspot.com/-f4oZ5Lplb3A/UslR1aZuPLI/AAAAAAAA1F8/EQCPRA377R0/s1600/dogs-003-017.jpg'
r = requests.get(url)
filename = 'image.jpg'
with open(filename, 'wb') as out_file:
    out_file.write(r.content)
    
print('Download Completed')
```

```python
image = cv2.imread(filename)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.figure(figsize = (8,8))
plt.imshow(gray_image, cmap = "gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.show()
```

```python
inverted_image = cv2.bitwise_not(gray_image)
plt.figure(figsize = (8,8))
plt.imshow(inverted_image, cmap = "gray")
plt.axis("off")
plt.title("Inverted Image")
plt.show()
```

```python
image_smoothing = cv2.GaussianBlur(inverted_image, (21,21),sigmaX = 0, sigmaY = 0)
plt.figure(figsize = (8,8))
plt.imshow(image_smoothing, cmap = "gray")
plt.axis("off")
plt.title("Image Smoothing")
plt.show()
```

```python
pencil_sketch = cv2.divide(gray_image, 255 - image_smoothing ,scale = 255)
plt.figure(figsize = (8,8))
plt.imshow(pencil_sketch, cmap = "gray")
plt.axis("off")
plt.title("Pencil Sketch")
plt.show()
```
